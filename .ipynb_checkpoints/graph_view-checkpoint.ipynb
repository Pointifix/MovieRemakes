```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import json

with open('remakes_edges.csv', 'r') as csv:
    edges_table = pd.read_csv(csv, sep=";", header=[0]).drop_duplicates()
    
with open('remakes_nodes.csv', 'r') as csv:
    nodes_table = pd.read_csv(csv, sep=";", header=[0]).drop_duplicates()

print(edges_table)
print(nodes_table)
```

```python
plt.subplots(figsize=(20, 15))


nx_graph = nx.DiGraph()

nx_graph.add_nodes_from(nodes_table["name"])

for entry in edges_table.iterrows():    
    nx_graph.add_edge(nodes_table["name"][entry[1]["director_original"] - 1],
                      nodes_table["name"][entry[1]["director_remake"] - 1])

nx.draw_circular(nx_graph)
```

```python
cola_js_string = '{\n\t"nodes":[\n'

movies_original_dict = {}
counter_original = 0

for remake in edges_table.iterrows():    
    original_title = remake[1].title_original    
    original_director_id = remake[1].director_original
    
    if original_title not in movies_original_dict:
        movies_original_dict[original_title] = counter_original
    
        cola_js_string += '\t\t{"name": "' + original_title + '", "group": ' + str(original_director_id) + '},\n'
        counter_original += 1

movies_remake_dict = {}        
counter_remake = counter_original

for remake in edges_table.iterrows():
    remake_title = remake[1].title_remake    
    remake_director_id = remake[1].director_remake
    
    if remake_title not in movies_remake_dict:
        movies_remake_dict[remake_title] = counter_remake
    
        cola_js_string += '\t\t{"name": "' + remake_title + '", "group": ' + str(remake_director_id) + '},\n'
        counter_remake += 1
    
print(movies_original_dict)

print(movies_remake_dict)
    
cola_js_string = cola_js_string[:-2]
cola_js_string += '\n\t],\n\t"links":[\n'

for remake in edges_table.iterrows():
    film_original = remake[1].title_original
    film_remake = remake[1].title_remake
    
    cola_js_string += '\t\t{"source": ' + str(movies_original_dict[film_original]) + ',"target": ' + str(movies_remake_dict[film_remake]) + ',"value": 1},\n'

cola_js_string = cola_js_string[:-2]
cola_js_string += '\n\t]\n}'

print(cola_js_string)
```

```python
#output for arc diagram d3
movies_by_year = []
remakes = []

def get_movie_by_year(movie):
    return movie.get("year")

original_movies = {}

id_counter = 0

for remake_entry in edges_table.iterrows():
    original_movie = remake_entry[1].title_original
    original_year = remake_entry[1].year_original
    
    id_original = id_counter
    if original_movie not in original_movies:
        original_movies[original_movie] = id_original
        movies_by_year.append({"name": original_movie, "year": original_year, "id": id_original, "original": True})
        id_counter += 1
    else:
        id_original = original_movies[original_movie]

    remake_movie = remake_entry[1].title_remake
    remake_year = remake_entry[1].year_remake
    
    id_remake = id_counter
    movies_by_year.append({"name": remake_movie, "year": remake_year, "id": id_remake, "original": False})
    id_counter += 1
    
    remakes.append((id_original, id_remake))

movies_by_year.sort(key=get_movie_by_year)

for i, remake in enumerate(remakes):
    original_id = [i for i,movie in enumerate(movies_by_year) if movie.get("id") == remake[0]][0]
    remake_id = [i for i,movie in enumerate(movies_by_year) if movie.get("id") == remake[1]][0]
    
    remakes[i] = (original_id, remake_id)
    
# generate output string
d3_string = '{\n\t"nodes":[\n'

for i, movie in enumerate(movies_by_year):
    d3_string += '\t\t{"id": ' + str(i) + ', "name": "' + movie.get("name") + '", "year": ' + str(movie.get("year")) + ', "original": ' + str(movie.get("original")).lower() + '},\n'
    
d3_string = d3_string[:-2]
d3_string += '\n\t],\n\t"links":[\n'
    
for remake in remakes:
    d3_string += '\t\t{"source": ' + str(remake[0]) + ', "target": ' + str(remake[1]) + '},\n'
    
d3_string = d3_string[:-2]
d3_string += '\n\t]\n}'
    
print(d3_string)
```

```python
#output string custom
output_dict = {}

#directors
directors_list = []
for director_entry in nodes_table.iterrows():
    directors_list.append({"id": director_entry[1].id - 1, "name": director_entry[1][1]})
    
output_dict["directors"] = directors_list
    
#movies and links
unique_movies_dict = {}    
movies_list = []
links_list = []
id_counter = 0
for remake_entry in edges_table.iterrows():
    movies_dict_key = (remake_entry[1].title_original, remake_entry[1].year_original)
    if movies_dict_key not in unique_movies_dict:
        unique_movies_dict[movies_dict_key] = {"id": id_counter, "directors": [remake_entry[1].director_original - 1]})
        movies_list.append({"id": id_counter, 
                            "title": remake_entry[1].title_original, 
                            "year": remake_entry[1].year_original, 
                            "directors": [remake_entry[1].director_original - 1], 
                            "original": True})
        id_counter += 1
    elif remake_entry[1].director_original - 1 not in unique_movies_dict[movies_dict_key]["directors"]:
        unique_movies_dict[movies_dict_key]["directors"].append(remake_entry[1].director_original - 1)
        movies_list[unique_movies_dict[movies_dict_key]["id"]]["directors"].append(remake_entry[1].director_original - 1)
    id_original = unique_movies_dict[movies_dict_key]
    
    movies_dict_key = (remake_entry[1].director_remake, remake_entry[1].title_remake, remake_entry[1].year_remake)
    if movies_dict_key not in unique_movies_dict:
        unique_movies_dict[movies_dict_key] =  {"id": id_counter, "directors": [remake_entry[1].director_remake - 1]})
        movies_list.append({"id": id_counter, 
                            "title": remake_entry[1].title_remake, 
                            "year": remake_entry[1].year_remake, 
                            "directors": [remake_entry[1].director_remake - 1], 
                            "original": False})
        id_counter += 1
    elif remake_entry[1].director_remake - 1 not in unique_movies_dict[movies_dict_key]["directors"]:
        unique_movies_dict[movies_dict_key]["directors"].append(remake_entry[1].director_remake - 1)
        movies_list[unique_movies_dict[movies_dict_key]["id"]]["directors"].append(remake_entry[1].director_remake - 1)
    id_remake = unique_movies_dict[movies_dict_key]
        
    links_list.append({"id_original": id_original, "id_remake": id_remake})

output_dict["movies"] = movies_list
output_dict["links"] = links_list
        
obj = json.dumps(output_dict, indent = 2)
print(obj)
```
